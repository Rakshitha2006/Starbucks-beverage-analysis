```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv("starbucks.csv")
df
```

```python
df.isnull()
```

```python
count=df["Beverage_category"].value_counts()
print(count)
```

```python
beverage_unique=df["Beverage_category"].unique()
beverage_unique
```

```python
x=beverage_unique
y=count
plt.figure(figsize=(10,6))
plt.barh(x,y,color="g")
plt.title("Beverages categories count")
plt.show()
```

```python
average_calories_per_category = df.groupby('Beverage_category')['Calories'].mean()
plt.figure(figsize=(28,6))
plt.plot(beverage_unique,average_calories_per_category,color="r",marker="o")
plt.title("Average calories for each category")
plt.show()
```
