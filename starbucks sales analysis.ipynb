```python
#importing the libraries and data
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv("starbucks.csv")
df
```

```python
df.info()
```

```python
#To check whether it has null values
df.isnull().sum()
```

```python
#Handling the null values using fillna
df["Caffeine (mg)"].fillna('125', inplace=True)
```

```python
df.isnull().sum()
```

```python
#To check the datatypes
df.dtypes
```

```python
#To show the column names of the dataset
df.columns
```

```python
#To show the number of rows and columns in the DataFrame
df.shape
```

```python
df.nunique() #To show the number of unique values in each column
```

```python
x=df["Beverage_category"].unique()
y=df["Beverage_category"].value_counts()
sort_count=y.sort_values()
plt.figure(figsize=(10,6))
plt.barh(x,sort_count,color="brown")
plt.title("Count of each beverage category")
plt.xlabel("Count")
plt.ylabel("Beverages category")
plt.show()
```

```python
x=df["Beverage_category"].unique()
average_calories_per_category = df.groupby('Beverage_category')['Calories'].mean()
plt.figure(figsize=(28,6))
plt.plot(x,average_calories_per_category,color="r",marker="o")
plt.xlabel("Beverage category")
plt.ylabel("Calories")
plt.title("Average calories for each category")
plt.show()
```

```python
plt.figure(figsize=(17, 5))
sns.histplot(df.iloc[:,-1], kde=True,color='g')
plt.title('Caffeine Distribution')
plt.xlabel('Caffeine')
plt.show()
```

```python
plt.figure(figsize=(12, 5))
sns.histplot(df.iloc[:,4], kde=True, label='Total Fat', color='yellow')
sns.histplot(df.iloc[:,5], kde=True, label='Trans Fat', color='red')
plt.xlabel('Fat Content (g)')
plt.ylabel('Frequency')
plt.title('Comparison of Total Fat and Trans Fat')
plt.legend()
plt.show()
```

```python
prep_calories = df.groupby('Beverage_prep')['Calories'].mean()
top3_prep_calories = prep_calories.sort_values(ascending=False).head(3)
plt.figure(figsize=(10, 6))
top3_prep_calories.plot(kind='bar', color='pink')
plt.xlabel('Beverages')
plt.ylabel('Calories')
plt.title('Top 3 beverages with most calories')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.boxplot(x='Beverage_category', y='Calories', data=df)
plt.title('Calories distribution for each beverage category')
plt.xlabel('Beverage Category')
plt.ylabel('Calories')
plt.xticks(rotation=45, ha='right') # X axis rotation
plt.show()
```

```python
x=df.iloc[:,0]
y1=df.iloc[:,8]
plt.figure(figsize=(25,5))
plt.bar(x,y1,color='m')
plt.xlabel("Beverage_Category")
plt.ylabel(" Total Carbohydrates")
plt.title("Total Carbohydrates for each Beverage Category")
plt.show()
```

```python
cat_sugar = df.groupby('Beverage')[' Sugars (g)'].mean()
sorted_cat_sugar = cat_sugar.sort_values(ascending=False)
plt.figure(figsize=(10, 3))
sorted_cat_sugar.plot(kind='bar', color='orange')
plt.xlabel('Beverage Category')
plt.ylabel('Total Sugar (g)')
plt.title('Average Sugar in each Beverages')
plt.show()
```

```python
x=df.iloc[:,7].unique()
mylabels=df.iloc[:,0].unique()
plt.figure(figsize=(10,4))
plt.pie(x,labels=mylabels,autopct="%1.1f%%")
plt.title("Sodium distribution in each beverage category")
plt.show()
```

```python
x=df.iloc[:,0]
y1=df.iloc[:,-3]
plt.figure(figsize=(30,6))
plt.bar(x,y1,color="#D7B3DC")
plt.xlabel("Beverage_Category")
plt.ylabel("Calcium")
plt.title("Calcium in each Beverage Category")
plt.show()
```

```python
x = df.groupby('Beverage')['Cholesterol (mg)'].mean()
top5 = x.sort_values(ascending=False).head(5)
plt.figure(figsize=(10, 6))
top5.plot(kind='bar', color='yellow')
plt.xlabel('Beverages')
plt.ylabel('Cholestrol')
plt.title('Top 5 beverages with more cholestrol')
plt.show()
```

```python
df.columns = df.columns.str.strip()
trans_fat_by_category = df.groupby('Beverage_category')['Saturated Fat (g)'].sum()
plt.figure(figsize=(15,6))
plt.pie(trans_fat_by_category, labels=trans_fat_by_category.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Saturated Fat by Beverage Category ')
plt.show()
```

```python
df.columns = df.columns.str.strip()
protein_by_category = df.groupby('Beverage_category')['Protein (g)'].sum()
plt.figure(figsize=(10, 8))
plt.pie(protein_by_category, labels=protein_by_category.index, autopct='%1.1f%%', startangle=90)
plt.title('Distribution of Protein by Beverage Category ')
plt.show()
```
